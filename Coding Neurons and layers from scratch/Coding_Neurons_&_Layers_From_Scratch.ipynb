# Layer of Neurons 

![My image](Layer_of_Neurons.jpg)

```python
inputs = [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

weights1 = weights[0] #LIST OF WEIGHTS ASSOCIATED WITH 1ST NEURON : W11, W12, W13, W14
weights2 = weights[1] #LIST OF WEIGHTS ASSOCIATED WITH 2ND NEURON : W21, W22, W23, W24
weights3 = weights[2] #LIST OF WEIGHTS ASSOCIATED WITH 3RD NEURON : W31, W32, W33, W34

biases = [2, 3, 0.5]

bias1 = 2
bias2 = 3
bias3 = 0.5

outputs = [
    #Neuron_1
    inputs[0]*weights1[0]+
    inputs[1]*weights1[1]+
    inputs[2]*weights1[2]+
    inputs[3]*weights1[3] + bias1,
    
    #Neuron_2
    inputs[0]*weights2[0]+
    inputs[1]*weights2[1]+
    inputs[2]*weights2[2]+
    inputs[3]*weights2[3] + bias2,
    
    #Neuron_3
    inputs[0]*weights3[0]+
    inputs[1]*weights3[1]+
    inputs[2]*weights3[2]+
    inputs[3]*weights3[3] + bias3
]

print(outputs)
```

# USING LOOPS FOR BETTER AND EASIER CODING

```python
# zip() is a Python function that pairs things together â€” item-by-item
a = [1, 2, 3]
b = ['a', 'b', 'c']

list(zip(a, b))
```

```python
inputs = [1, 2, 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1],  # 1 neuron have 4 weights 
 [0.5, -0.91, 0.26, -0.5],
 [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

outputs =[]

for neuron_weight, neuron_bias in zip(weights,biases):
    neuron_output = 0
    
    for x,w in zip(inputs,neuron_weight):
        neuron_output += x * w
        
    neuron_output += neuron_bias
    outputs.append(neuron_output)
    
print(outputs)
```
